```python
nameList = ['유나', '지은', '스튜어트', '케빈']
sentence = '우리 강아지는 쇼파 위에서 잔다.'
```

nameList : str 객체의 리스트<br>
sentence : str 객체인 하나의 문장

```python
names = ':'.join(nameList)
print(type(names),':',names)
```

```python
wordList = sentence.split(' ')
print((type(wordList)), ':', wordList)
```
