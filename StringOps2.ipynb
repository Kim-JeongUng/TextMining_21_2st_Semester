```python
str = 'NLTK Dolly Python'
```

```python
print(str)
```

```python
print('다음의 인덱스에서 끝나는 부분 문자열 :', str[:4])
```

```python
print('다음의 인덱스에서 끝나는 부분 문자열 :', str[11:])
```

```python
print('부분 문자열 :', str[5:10])
```

```python
print('부분 문자열 :', str[5:11])
```

```python
print('복잡한 방식의 부분 문자열 :', str[-12:-7])
```

```python
if 'NLTK' in str:
  print('NLTK를 찾았습니다')
```

```python
replaced = str.replace('Dolly', 'Dorothy')
print('대체된 문자열', replaced)
```

```python
print('각 문자(chracter) 액세스 : ')
for s in replaced:
  print(s)
```

```python
print('각 문자(chracter) 액세스 : ')
for s in str:
  print(s)
```

string 객체 : 문자(character)의 리스트   
for구문을 사용하여 모든 문자에 접근 가능   
':' in list : 리스트의 일부 리턴   
':' 다음 숫자 : 0 부터 인덱스에서 을 뺀 값으로 끝나는 부분 리스트 ':'이전 숫자 : 주어진 숫자부터 끝까지 리스트
