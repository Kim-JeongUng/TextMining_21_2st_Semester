```python
import nltk, matplotlib, wordcloud
```

```python
import matplotlib.pyplot as plt
```

```python
from nltk.corpus.reader.plaintext import PlaintextCorpusReader
```

```python
from wordcloud import WordCloud
```

```python
from nltk.tokenize import LineTokenizer, SpaceTokenizer, TweetTokenizer
```

```python
from nltk import word_tokenize
```

```python
nltk.download('punkt')
```

```python
newCorpus = PlaintextCorpusReader('/content/drive/MyDrive/Colab Notebooks', 'simple_corpus.txt')
```

```python
sim_words = newCorpus.words()
fdist = nltk.FreqDist(sim_words)
```

```python
print('말뭉치에서 가장 흔한 15개 단어는 다음과 같습니다')
print(fdist.most_common(15))
```

```python
fdist.plot(cumulative=True)
```

```python
clean_words = dict([(m, n) for m, n in fdist.items() if len(m) > 3])

wordcloud = WordCloud().generate_from_frequencies(clean_words)

plt.figure(figsize=(12,9))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
